### Finding the missing values and handling null values

```python
df.isna().sum()
```

If the percentage of null values is greater than 70 percent,then we can delete respective column.<br>As the dataset consists of minimum null values,We can replace the null values based on their distribution.<br> 1) If their is normal distribution then null values are replaced with mean of the feature.<br> 2) If the distribution is skewed then null values are replaced with median of the feature  

converting "Total_Sales" column to numeric in order to find distribution ,the null values in column are strings.So filling null strings with NA and converting into numeric

```python
df['Total_Sales']=df['Total_Sales'].replace(" ",np.nan)
```

```python
## Converting 'Total_Sales' to numeric from Object type
df['Total_Sales'] = pd.to_numeric(df['Total_Sales'])
```

```python
df['Total_Sales'].dtype
```

Datatype Conversion completed

#### Finding distribution of the features to replace null values

```python
sns.displot(x='Total_Sales',data=df)
```

Distribution plots shows<br>
1)Base_pay feature is normally distributed ,so replacing Null values with mean<br>
2)openingbalance feature is right skewed,so replacing Null values with median <br>
3)Total_Sales feature is right skewed,so replacing Null values with median

```python
df['Base_pay']=df['Base_pay'].replace(np.nan,df['Base_pay'].mean())
df['Total_Sales']=df['Total_Sales'].replace(np.nan,df['Total_Sales'].median())
df['openingbalance']=df['openingbalance'].replace(np.nan,df['openingbalance'].median())
```

## Outliers handeling

```python
df.boxplot('Base_pay')
```

```python
#  IQR method:
q1 = df.Base_pay.quantile(0.25)
q3 = df.Base_pay.quantile(0.75)
IQR = q3-q1
print("IQR: ",IQR)
lower_bound = q1-(1.5*IQR)
upper_bound = q3+(1.5*IQR)
print("lower bound=",lower_bound)
print("upper bound=",upper_bound)
```

```python
df = df[(df.Base_pay>lower_bound) & (df.Base_pay<upper_bound)]
df.shape
```

I am handelling the outliers of BasePay as it is highly correlated and has NA values and removing outliers in Basepay,I noticed that many outliers are reduced in categorical features and it seems me a good idea.

### Linear Regression

```python
from sklearn import linear_model, metrics

X = df1.drop(columns=['Salary'])
y = df1['Salary']

# splitting X and y into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=22)

reg = linear_model.LinearRegression()

# train the model using the training sets
reg.fit(X_train, y_train)

y_prediction =  reg.predict(X_test)
y_prediction

# importing r2_score module
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
# predicting the accuracy score
score=r2_score(y_test,y_prediction)
print("r2 score = ",score)
print("Mean_squared_error = ",mean_squared_error(y_test,y_prediction))
print("Root_Mean_Squared Error = ",np.sqrt(mean_squared_error(y_test,y_prediction)))
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)


y_pred = regressor.predict(X_test)

from sklearn import metrics

from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
# predicting the accuracy score
score=r2_score(y_test,y_pred)
print("r2 score = ",score)
print("Mean_squared_error = ",mean_squared_error(y_test,y_pred))
print("Root_Mean_Squared Error of = ",np.sqrt(mean_squared_error(y_test,y_pred)))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

X = df1.drop(columns=['Salary'])
y = df1['Salary']

# splitting X and y into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=1)

# create linear regression object

rf=RandomForestRegressor(n_estimators=1000,random_state=42)

# train the model using the training sets
rf.fit(X_train, y_train)
y_prediction =  rf.predict(X_test)
y_prediction

from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
# predicting the accuracy score
score=r2_score(y_test,y_prediction)
print("r2 score = ",score)
print("Mean_squared_error = ",mean_squared_error(y_test,y_prediction))
print("Root_Mean_Squared Error = ",np.sqrt(mean_squared_error(y_test,y_prediction)))
```
